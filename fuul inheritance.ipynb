```python
#METHOD RESOULTION
class batman:
    def __init__ (self):
             self.heroname = input("enter heronam:")
             self.studio =input("enter studio name:")
    def view(self):
        print("heroname:",self.heroname)
        print("studio belongs:",self.studio)
        
class  Child:
     def __init__(self):
            self.heroname =input("enter heronam:")
            self.studio =input("enter studio name:")
            
     def show(self):
             print("heroname:",self.heroname)
             print("studio belongs:",self.studio)
        
class grandChild(batman , Child):
        def __init__(self):
            super().__init__()
            self.fheroname = input("enter heronam:")
            self.fstudio =input("enter studio name:")
        def output(self):
                super().view()
                print("heroname:",self.fheroname)
                print("studio belongs:",self.fstudio) 
            

ob1 = grandChild()
ob1.output()            
```

```python
#Hierarchical Inheritance
class batman:
    def __init__ (self):
             self.heroname = input("enter heronam:")
             self.studio =input("enter studio name:")
    def view(self):
        print("heroname:",self.heroname)
        print("studio belongs:",self.studio)
        
class  Child (batman):
     def __init__(self):
            self.heroname =input("enter heronam:")
            self.studio =input("enter studio name:")
            
     def show(self):
        print("heroname:",self.heroname)
        print("studio belongs:",self.studio)
        
class grandChild(batman):
        def __init__(self):
            self.fheroname = input("enter heronam:")
            self.kstudio =input("enter studio name:")
        def output(self):
            print("heroname:",self.fheroname)
            print("studio belongs:",self.kstudio) 

obj=batman()
obj1=Child()
obj.view()
obj1.show()
obj2= grandChild()
obj2.output()
```

```python
#hybrid
class batman:
    def __init__ (self):
             self.heroname = input("enter heronam")
             self.studio =input("enter studio name")
    def view(self):
        print("heroname:",self.heroname)
        print("studio belongs:",self.studio)
        
class  Child ():
     def __init__(self):
            parent.__init__ (self)
            self.sheroname =input("enter heronam")
            self.hstudio =input("enter studio name")
     def show(self):
        print("heroname:",self.sheroname)
        print("studio belongs:",self.hstudio)
class  grandChild ():
     def __init__(self):
            Child.__init__ (self)
            self.sheroname =input("enter heronam")
            self.hstudio =input("enter studio name")
     def show(self):
        print("heroname:",self.sheroname)
        print("studio belongs:",self.hstudio)        
        
class greatgrandChild(Child , grandChild):
        def __init__(self):
            grandChild.__init__ (self)
            self.nheroname = input("enter heronam")
            self.mstudio =input("enter studio name")
        def output(self):
            print("heroname:",self.nheroname)
            print("studio belongs:",self.mstudio) 
        
obj=greatgrandChild()
obj.output()
```
